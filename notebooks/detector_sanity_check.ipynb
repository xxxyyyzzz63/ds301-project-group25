```python
import os
import shutil

# make folders
os.makedirs("models", exist_ok=True)
os.makedirs("src", exist_ok=True)
os.makedirs("notebooks", exist_ok=True)

# move files if they exist
file_moves = {
    "diverse_week5_artifacts.pkl": "models/diverse_week5_artifacts.pkl",
    "final_detector.py": "src/final_detector.py",
    "stylometry_features.py": "src/stylometry_features.py",
    "test_final_detector.py": "test_final_detector.py",  # keep in repo root
}

for src, dst in file_moves.items():
    if os.path.exists(src):
        shutil.move(src, dst)
        print(f"Moved: {src} -> {dst}")
    else:
        print(f"Not found: {src}")

print("\nDone.")
print("Root contents:", os.listdir("."))
print("models contents:", os.listdir("models"))
print("src contents:", os.listdir("src"))
```
